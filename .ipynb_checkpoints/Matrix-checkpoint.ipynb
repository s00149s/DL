### 출력층

● 은닉층 활성화 함수 : 시그모이드  
● 출력층 활성화 함수 : 항등 함수  
● 손실 함수 : 오차 제곱합  
● 최적화 알고리즘 : SGD(input에서 batch 내용을 random하게 섞어서 뽑아내는 알고리즘)  
● 배치 사이즈 : 1  

```python
# 출력층 class 정의 
class OutputLayer:
    # 생성자
    def __init__(self, n_upper, n):
        # n_upper: 가중치의 로우 수(입력의 컬럼 수) = 이전 레이어의 노드 갯수
        # n : 출력 갯수(1), 바이어스 = 출력갯수
        self.w = wb_width * np.random.randn(n_upper, n) 
        self.b = wb_width * np.random.randn(n)
        # wb_width = 0.01 가중치와 편향 설정을 위한 정규분표의 표준편차
        
    def forward(self, x):
        self.x = x
        u = np.dot(x, self.w) + self.b
        self.y = u # 최종 출력의 activation 함수는 항등함수 사용
        
    def backward(self, t):
        delta = self.y - t
        # 가중치 기울기 
        self.grad_w = np.dot(self.x.T, delta)
        # 바이어스 기울기
        self.grad_b = np.sum(delta, axis=0)
        # 출력 기울기
        self.grad_x = np.dot(delta, self.w.T)
        
    # 가중치와 bias를 수정하는 방법(기울기 0인 곳 찾아가기)
    def update(self, eta):
        self.w -= eta * self.grad_w
        self.b -= eta * self.grad_b
```

### 은닉층

```python
class MiddleLayer:
    def __init__(self, n_upper, n):
        # randn : 0에서 1사이의 난수 생성
        self.w = wb_width * np.random.randn(n_upper, n) 
        self.b = wb_width * np.random.randn(n)
        
    def forward(self, x):
        self.x = x
        u = np.dot(x, self.w) + self.b
        self.y = 1 / (1 + np.exp(-u)) # 시그모이드 함수
        
    def backward(self, grad_y):
        delta = grad_y * (1 - self.y) * self.y# 시그모이드 미분
        self.grad_w = np.dot(self.x.T, delta)
        self.grad_b = np.sum(delta, axis=0)
        self.grad_x = np.dot(delta, self.w.T)
        
    def update(self, eta):
        self.w -= eta * self.grad_w
        self.b -= eta * self.grad_b
```

#### 메인코드

```python
import matplotlib.pyplot as plt
```

```python
# input data 정의하기
input_data = np.arange(0, np.pi*2, 0.1)
# 정답
correct_data = np.sin(input_data)
input_data = (input_data - np.pi) / np.pi # 입력을 -1.0 ~ 1.0 넣어주기 위해서 변형
n_data = len(correct_data) # 전체 데이터의 갯수

# 초기값 설정
n_in = 1 # 입력층의 뉴런 수 
n_mid = 5 # 은닉층의 뉴런 수 
n_out = 1 # 출력층의 뉴런 수 

wb_width = 0.01
eta = 0.1 # 학습률 learning rate
epoch = 2001 # 에포크 2000번 돌리기
interval = 200 # epoch 1마다 얼마나 진행되고 있는지 찍어보기

# 각 층의 초기화
middle_layer = MiddleLayer(n_in, n_mid)
output_layer = OutputLayer(n_mid, n_out)

# 학습시키기
for i in range(epoch):    
    # number of data
    index_random = np.arange(n_data)
    # index 섞어주기
    np.random.shuffle(index_random)
    
    # 결과 표시할 변수 정하기
    total_error = 0
    plot_x = []
    plot_y = []
    
    # 섞은 인덱스에서 함수 뽑아내기
    for idx in index_random:
        # 무작위로 나온 것 중에서 x값과 target을 매칭시켜줌 
        x = input_data[idx:idx+1]
        t = correct_data[idx:idx+1]
        
        # 순전파
        middle_layer.forward(x.reshape(1,1)) # input을 행렬로 변환
        output_layer.forward(middle_layer.y) # middle_layer에서 나온 출력값 입력해주기
        
        # 역전파
        output_layer.backward(t.reshape(1,1))
        middle_layer.backward(output_layer.grad_x) # outputlayer에서 구한 grad_x
        
        # 가중치와 바이어스 학습률 넣고 수정
        middle_layer.update(eta)
        output_layer.update(eta)
        
        # 중간중간 진행상황 찍어주기
        # 학습과는 관련 없는 코드
        if i % interval == 0:
            y = output_layer.y.reshape(-1) # 2차원값을 1차원으로 바꿔줌 -> 갯수는 안정해줘도 지정됨 
            
            total_error += 1.0 / 2.0 * np.sum(np.square(y - t)) # 토탈 에러 누적
            
            plot_x.append(x)
            plot_y.append(y)
            
    # 그래프 그리기
    if i % interval == 0:
        plt.plot(input_data, correct_data, linestyle="dashed")
        plt.scatter(plot_x, plot_y, marker='+')
        plt.show()
        
        print("Epoch:"+str(i)+'/'+str(epoch),\
              "Error:"+str(total_error/n_data))
    
```

- 은닉층의 뉴런수를 1, 2, 3, 4, 5 차례로 증가시키며 결과를 확인  
- 중간에 있는 Hidden Layer의 뉴런 수가 결과에 영향을 미치는가?  
- 보통은 뉴런 수가 늘어날수록 과대적합쪽으로 가지만, 현재는 사인 그래프가 단순해서 크게 변화 없음
- 1 : 0.28 / 0.03  => 과소적합
- 2 : 0.27 / 0.01  
- 3 : 0.27 / 8.58  
- 4 : 0.28 / 9.83
- 5 : 0.29 / 7.79
- 3 또는 4개의 뉴런 이후에는 결과가 더 좋아지지 않고 계산량만 증가  
- 은닉층에서 필요 이상의 뉴런 수는 과적합(Overfitting)을 일으킴
